```python
# import library
import pandas as pd
import numpy as np
```

```python
# untuk mendapatkan dataset latih dari github
!wget -O diabetes_latih.csv https://raw.githubusercontent.com/thea725/orbit-archive/main/datasets/diabetes_latih.csv
```

```python
# untuk mendapatkan dataset uji dari github
!wget -O diabetes_uji.csv https://raw.githubusercontent.com/thea725/orbit-archive/main/datasets/diabetes_uji.csv
```

```python
# membaca file diabetes_latih.csv sebagai csv
df = pd.read_csv('diabetes_latih.csv')

X_train = df.values
X_train = np.delete(X_train,8, axis=1)

y_train = df['Outcome'].values

df = pd.read_csv('diabetes_uji.csv')

X_test = df.values
X_test = np.delete(X_test,8,axis=1)

y_test = df['Outcome'].values
```

```python
from sklearn.preprocessing import MinMaxScaler
X_train = MinMaxScaler().fit_transform(X_train)
X_test = MinMaxScaler().fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
# mendefinisikan model
model = Sequential()

model.add(Dense(12, activation='relu', input_shape=(8,)))

model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# menampilkan dataset
df.head()
```

```python
model.fit(X_train, y_train, epochs=200, verbose=1, batch_size=20)

y_pred = model.predict(X_test)
y_pred = y_pred >= 0.13
```

```python
# menampilkan skor akurasi
from sklearn.metrics import accuracy_score

round(accuracy_score(y_test, y_pred),3)
```
